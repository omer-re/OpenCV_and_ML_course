```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import time
```

```python
nadia = cv2.imread('Computer-Vision-with-Python/DATA/Nadia_Murad.jpg',0)
denis = cv2.imread('Computer-Vision-with-Python/DATA/Denis_Mukwege.jpg',0)
solvay = cv2.imread('Computer-Vision-with-Python/DATA/solvay_conference.jpg',0)

plt.imshow(nadia,cmap='gray')
```

```python
## CLASSIFIER
face_cascade = cv2.CascadeClassifier('Computer-Vision-with-Python/DATA/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    
    face_img=img.copy()
    face_rects = face_cascade.detectMultiScale(face_img)
    cnt=10
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(0,255,0),10)
        #cv2.putText(face_img, text='Reps count:'+cnt, org=(50,400), fontFace=font,fontScale=2,color=(255,255,255), thickness=3, lineType=cv2.LINE_AA)

        if y>200:
            cnt=cnt-1
            #time.wait(1000)
        
    return face_img
```

```python
result = detect_face(solvay)
plt.imshow(result,cmap='gray')
```

```python
cap = cv2.VideoCapture(0) 

while True: 
    
    ret, frame = cap.read(0) 
     
    frame = detect_face(frame)
    
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.line(frame,pt1=(0,200),pt2=(700,200), color=(0,0,255), thickness=5)

    #cv2.putText(frame, text='Reps count: 10', org=(50,400), fontFace=font,fontScale=2,color=(255,255,255), thickness=3, lineType=cv2.LINE_AA)
    cv2.imshow('Video Face Detection', frame) 
 
    c = cv2.waitKey(1) 
    if c == 27: 
        break 
        
cap.release() 
cv2.destroyAllWindows()
```
