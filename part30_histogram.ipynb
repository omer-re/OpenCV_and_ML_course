```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dark_horse = cv2.imread('Computer-Vision-with-Python/DATA/horse.jpg')
show_horse = cv2.cvtColor(dark_horse, cv2.COLOR_BGR2RGB)

rainbow = cv2.imread('Computer-Vision-with-Python/DATA/rainbow.jpg')
show_rainbow =cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)

blue_bricks = cv2.imread('Computer-Vision-with-Python/DATA/bricks.jpg')
show_bricks = cv2.cvtColor(blue_bricks, cv2.COLOR_BGR2RGB)
```

```python
hist_values =cv2.calcHist([blue_bricks], channels = [0], mask=None, histSize=[256], ranges=[0,256])
hist_values.shape
```

```python
plt.plot(hist_values)
```

```python
hist_values2 =cv2.calcHist([dark_horse], channels = [0], mask=None, histSize=[256], ranges=[0,256])
hist_values2.shape
plt.plot(hist_values2)
```

```python
img = blue_bricks
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i], None,[256],[0,256])
    plt.plot(histr, color=col)
    
plt.title('HISTOGRAM FORBLUE BRICKS')
```

```python
rainbow = cv2.imread('Computer-Vision-with-Python/DATA/rainbow.jpg')
show_rainbow =cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)
img=rainbow
img.shape
```

```python
mask = np.zeros(img.shape[:2],np.uint8)
plt.imshow(mask,cmap='gray')
mask[300:400,100:400]=255
```

```python
plt.imshow(mask,cmap='gray')
masked_imp = cv2.bitwise_and(img,img, mask=mask)
show_masked_img = cv2.bitwise_and(show_rainbow,show_rainbow,mask=mask)
plt.imshow(show_masked_img)
```

```python
hist_mask_values_red = cv2.calcHist([rainbow],channels=[2], mask=mask, histSize=[256],ranges=[0,256])
hist_values_red = cv2.calcHist([rainbow],channels=[2], mask=None, histSize=[256],ranges=[0,256])
plt.plot(hist_mask_values_red)
plt.title('RED HIST')
plt.plot(hist_values_red)
plt.title('RED HIST no mask')
```
