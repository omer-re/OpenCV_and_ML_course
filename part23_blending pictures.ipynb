```python
img1 = cv2.imread('Computer-Vision-with-Python/DATA/dog_backpack.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('Computer-Vision-with-Python/DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img2 = cv2.resize(img2,(600,600))
```

```python
plt.imshow(img2)
```

```python
img1.shape
```

```python
topleft_x = 934-600
topleft_y = 1401-600

bottomright_x = topleft_x + small_img.shape[1]
bottomright_y = topleft_y + small_img.shape[0]
```

```python
rows, cols, channels = img2.shape
```

```python
roi = img1[topleft_y:1401, topleft_x:943]
plt.imshow(roi)
```

```python
img2gray = cv2.cvtColor(img2,cv2.COLOR_RGB2GRAY)
plt.imshow(img2gray,cmap='gray')
```

```python
mask_inv = cv2.bitwise_not(img2gray)
white_bg = np.full(img2.shape,255,dtype=np.uint8)
white_bg.shape

#plt.imshow(mask_inv,cmap='gray')
```

```python
bk = cv2.bitwise_or(white_bg, white_bg, mask=mask_inv)
bk.shape
```

```python
plt.imshow(bk)
```

```python
fg = cv2.bitwise_or(img2,img2,mask=mask_inv)
plt.imshow(fg)
```

```python
final_roi = cv2.bitwise_or(roi,fg)
plt.imshow(final_roi)
```

```python
large_img=img1
small_img=final_roi
plt.imshow(img1)
```

```python
large_img[topleft_y:topleft_y+small_img.shape[0],topleft_x:topleft_x+small_img.shape[1]]= small_img
plt.imshow(large_img)
```
